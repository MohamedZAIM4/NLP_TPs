```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
nltk.download('stopwords')
```

```python
from google.colab import files
import io
uploaded = files.upload()
```

```python
# prompt: take the uploaded and store it in variable called text

text = uploaded['textTP2 (1)'].decode('utf-8')

print(text)
```

```python
# Prétraitement du texte
def preprocess_text(text):
    # Tokenization
    tokens = word_tokenize(text)
    # Retrait des mots vides et de la ponctuation
    stop_words = set(stopwords.words('english'))
    tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]
    return tokens
tokenized_text = preprocess_text(text)
```

```python
tokenized_text
```

```python
# Entraînement du modèle Word2Vec
model = Word2Vec(sentences=[tokenized_text], vector_size=100, window=5, min_count=1, workers=4)
# 1- Extraire la représentation vectorielle d'un mot
word_vector = model.wv['morocco']
print("Vector representation of 'morocco':", word_vector)
```

```python
# 2- Calculer la similarité entre deux mots
similarity_score = model.wv.similarity('morocco', 'marrakech')
print("Similarity between 'morocco' and 'marrakech':", similarity_score)
```

```python
# 3- Extraire les mots contextuels (les plus similaires) pour un mot central donné
similar_words = model.wv.most_similar('marrakech', topn=5)
print("Words most similar to 'marrakech':", similar_words)
```
